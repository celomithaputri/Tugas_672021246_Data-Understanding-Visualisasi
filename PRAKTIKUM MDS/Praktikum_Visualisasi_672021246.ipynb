Pie chart

```python
import matplotlib.pyplot as plt
rasa_es_krim = ('Matcha', 'Cokelat', 'Vanilla', 'Jeruk', 'Mangga')
banyaknya = (10, 20, 30, 40, 50)
warna = ('#CD5C5C', '#F08080', '#FA8072', '#E9967A', '#FFA07A')
target_penjualan = (0, 0, 0, 0.1 ,0)

plt.pie(
    banyaknya,
    labels=rasa_es_krim,
    autopct='%1.1f%%',
    colors = warna,
    explode = target_penjualan,
    shadow = True
)
plt.show()
```

Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

negara = ('Argentina', 'Bolivia', 'Brazil', 'Chili', 'Colombia', 'Ecuador', ' Falkland Island', 'Guyana', 'Paraguay', 'Peru', 'Suriname')
jum_pend = (46537943, 12857870, 215112794	, 19800000, 52954279, 17963593, 2967, 255506, 835283, 34925628, 638144)

df = pd.DataFrame({
    'Country':negara,
    'Population':jum_pend,
})
df.sort_values(by='Population', inplace=True)

x_coords = np.arange(len(df))
warna= ['#0000FF' for _ in range(len(df))]
warna[-2] = '#FF0000'
plt.figure(figsize = (20,10))
plt.bar(x_coords, df['Population'], tick_label=df['Country'], color = warna)
plt.xticks(rotation = 90)
plt.ylabel('Populasi (Juta Jiawa)')
plt.title('Jumlah Penduduk Negara-Negara Di Amerika Serikat')
plt.show()
```

Scatter Plot

```python
import matplotlib.pyplot as plt

lemon_diameter = [6.44, 6.87, 7.7, 8.85, 8.15, 9.96, 7.21, 10.04, 10.2, 11.06]
lemon_berat = [112.05, 114.58, 116.71, 117.4, 128.93, 132.92, 138.93, 145.98, 148.44, 152.81]

lime_diameter = [6.15, 7.0, 7.0, 7.69, 7.95, 7.51, 10.46, 8.72, 9.53, 10.09]
lime_berat = [112.76, 125.18, 131.36, 132.41, 138.08, 142.55, 156.86, 158.67, 163.28, 166.74]

plt.scatter(lemon_diameter, lemon_berat, color='orange')
plt.scatter(lime_diameter, lime_berat, color='g')

plt.title('Lemon VS Lime')
plt.xlabel('Diameter (cm)')
plt.ylabel('Berat (gram)')
plt.legend(['lemon', 'lime'])
plt.show()
```
