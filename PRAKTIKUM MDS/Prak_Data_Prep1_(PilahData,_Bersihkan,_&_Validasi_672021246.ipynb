```python
import pandas as pd
import numpy as np

df = pd.read_csv("BL-Flickr-Images-Book.csv")

df.head()
```

```python
#untuk lihat tipe data masing-masing kolom/fitur
print(df.dtypes)
```

```python
#melihat statistika sederhana
df.describe()
```

```python
#hapus kolom
to_drop = ['Edition Statement','Corporate Author', 'Corporate Contributors', 'Former owner','Engraver','Issuance type', 'Shelfmarks']
df.drop(to_drop, inplace=True, axis = 1)
df.head()
```

```python
#Setting index
df['Identifier'].is_unique
```

```python
df = df.set_index('Identifier')
df.head()
```

```python
#untuk melihat isi data
df.loc[480]
```

```python
#merubah Date nya
df.loc[1905:, 'Date of Publication'].head(10)
```

```python
regex = r'^(\d{4})'
```

```python
extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand = False)
```

```python
extr.head()
```

```python
df['Date of Publication'] = pd.to_numeric(extr)
df['Date of Publication'].dtype
```

```python
df['Date of Publication'].isnull().sum() / len(df)
```

```python
df['Place of Publication'].head(20)
```

```python
df.loc[216]
```

```python
pub = df['Place of Publication']
london = pub.str.contains('London')
london [:5]
```

```python
oxford = pub.str.contains('Oxford')
```

```python
#panggil numpy where suruh perbaiki
df['Place of Publication'] = np.where(london, 'London', np.where(oxford, 'Oxford', pub.str.replace('-',' ')))
df['Place of Publication'].head()
```
