# **Soal 2: Permen**
Menggunakan Visualisasi "Pie Chart"

```python
import matplotlib.pyplot as plt

nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
warna = ('#CD5C5C', '#F08080', '#FA8072', '#E9967A', '#FFA07A')
jumlah_permen = [52, 39, 78, 13, 78]

#Mencari persentase peluang memilih permen Kopiko
peluang_kopiko = (jumlah_permen[nama_permen.index('Kopiko')] / sum(jumlah_permen)) * 100

#Membuat pie chart
plt.figure(figsize=(8, 8))
explode = (0, 0.1, 0, 0, 0)  #Menggunakan explode pada permen Kopiko
plt.pie(
    jumlah_permen,
    labels=nama_permen,
    autopct='%1.1f%%',
    colors = warna,
    explode=explode,
    shadow=True,
    startangle=140
)
plt.title('Persentase Peluang Memilih Permen Kopiko')
plt.show()

print("Peluang memilih permen Kopiko: {:.1f}%".format(peluang_kopiko))
```

# **Soal 3: Makanan**
Menggunakan Visualisasi "Bar Chart"

```python
import matplotlib.pyplot as plt

#Data makanan penutup
datapenjualan_makananpenutup = {
    'Donat': 14,
    'Pastel': 5,
    'Kue Coklat': 12,
    'Ice Cream': 19,
    'Puding Vanila': 6,
    'Brownies': 8,
    'Puding Strawberi': 12,
    'Puding Coklat': 9,
    'Ice Cream Nutela': 10,
    'Kue Coklat-Keju': 17,
    'Kue Wajik': 2,
    'Kue Sus': 9,
    'Mochi': 13,
}

#Mengurutkan makanan penutup berdasarkan jumlah penjualan
sorted_makananpenutup = sorted(datapenjualan_makananpenutup.items(), key=lambda x: x[1], reverse=True)

#Memisahkan nama makanan penutup dan jumlah penjualan
nama_makanan = [x[0] for x in sorted_makananpenutup]
jumlah_penjualan = [x[1] for x in sorted_makananpenutup]

#Membuat bar chart
plt.figure(figsize=(10, 6))
plt.barh(nama_makanan, jumlah_penjualan, color='skyblue')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Makanan Penutup')
plt.title('Popularitas Makanan Penutup')
plt.gca().invert_yaxis()
plt.show()

#Menentukan tiga makanan penutup yang paling tidak populer
makanan_tidak_populer = nama_makanan[-3:]
print("Makanan penutup yang harus dihapus:", makanan_tidak_populer)
```

1. Visualisasi yang digunakan adalah Bar Chart untuk menunjukkan popularitas relatif dari makanan penutup.
2. Makanan Penutup yang harus disingkirkan yaitu Puding Vanila, Pastel, Kue Wajik. Dapat dilihat dari jumlah penjualan.

# **Soal 4: Penggunaan CPU**
Menggunakan Visualisasi "Heatmap"

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Data penggunaan CPU
datapenggunaan_cpu = [
    [2, 2, 4, 2, 4, 1, 1, 4, 4, 12, 22, 23, 45, 9, 33, 56, 23, 40, 21, 6, 6, 2, 2, 3], # Senin
    [1, 2, 3, 2, 3, 2, 3, 2, 7, 22, 45, 44, 33, 9, 23, 19, 33, 56, 12, 2, 3, 1, 2, 2], # Selasa
    [2, 3, 1, 2, 4, 4, 2, 2, 1, 2, 5, 31, 54, 7, 6, 34, 68, 34, 49, 6, 6, 2, 2, 3], # Rabu
    [1, 2, 3, 2, 4, 1, 2, 4, 1, 17, 24, 18, 41, 3, 44, 42, 12, 36, 41, 2, 2, 4, 2, 4], # Kamis
    [4, 1, 2, 2, 3, 2, 5, 1, 2, 12, 33, 27, 43, 8, 38, 53, 29, 45, 39, 3, 1, 1, 3, 4], # Jumat
    [2, 3, 1, 2, 2, 5, 2, 8, 4, 2, 3, 1, 5, 1, 2, 3, 2, 6, 1, 2, 2, 1, 4, 3], # Sabtu
    [1, 2, 3, 1, 1, 3, 4, 2, 3, 1, 2, 2, 5, 3, 2, 1, 4, 2, 45, 26, 33, 2, 2, 1], # Minggu
]

#Label untuk sumbu x dan y
days = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']
hours = [str(i) + ':00' for i in range(24)]

#Membuat heatmap
plt.figure(figsize=(12, 7))
plt.title('Penggunaan CPU per Jam selama Seminggu')
plt.xlabel('Jam')
plt.ylabel('Hari')
sns.heatmap(datapenggunaan_cpu, cmap='YlGnBu', xticklabels=hours, yticklabels=days)
plt.show()
```

1. Jam berapa pekerja biasanya makan siang?
Jawab: Pada pukul 12:00, penggunaan CPU turun, menunjukkan waktu makan siang.

2. Apakah pekerja tersebut bekerja pada akhir pekan?
Jawab: Ya, mereka bekerja pada Sabtu dan Minggu. Terlihat dari heatmap bahwa penggunaan CPU pada akhir pekan tetap ada.

3. Pada hari apa pekerja mulai bekerja pada komputer mereka pada malam hari?
Jawab: Penggunaan CPU yang tinggi pada malam hari terjadi pada hari Kamis dan Jumat, menunjukkan bahwa pekerja mulai bekerja pada komputer mereka pada malam hari pada hari-hari tersebut.

Note: Warna dalam heatmap menunjukkan intensitas penggunaan CPU. Semakin gelap warna biru (dari warna hijau ke biru tua), semakin tinggi penggunaan CPU pada jam tersebut. Semakin terang warna biru (dari warna hijau ke biru muda), semakin rendah penggunaan CPU pada jam tersebut.

Dengan demikian, semakin berubahnya warna dari hijau ke biru tua menandakan peningkatan penggunaan CPU, sementara semakin terang warna biru menandakan penurunan penggunaan CPU.
